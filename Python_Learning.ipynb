### For loop practice:
- print even numbers from a given range in 1 to 10
### solution:
- An even number is something that can be divided by 2 and doesnt leave any reminder.
- In python how can it check if a number is even or it uses a modulo operator which basically divides the number by 2 and if doesnt leave any reminder then its even if it does leave reminder as 1 then its odd.
- now using a for loop we can check each number and calculate and if yes print the number if not move forward and check the other number.



```python
# Note:
#Even number = divisible by 2
#Check with % 2 == 0

for i in range(1,11):
  # i refers to index value range refers within a given limit of numbers
  if i%2 == 0:
    print(i)
  #else:
    #if i%2 != 0:
      #print(i)

```

```python
#Problem 2 : print number from a given range of 1 to 20 which are divisible by 3
#Pusedocode:
# given range is 1 to 20 (need to go through)
# need to divide each number by 3 and see if its leaving a reminder 0 or 1
#print all the number that are divisble by 3 using modulo operator.

for i in range(1,20):
  if i%3 == 0:
    print(i)
```
